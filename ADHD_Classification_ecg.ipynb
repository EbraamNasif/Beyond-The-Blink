```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

# this line allows the graphs to be displayed inside the jupyter notebook
%matplotlib inline
```

```python
ecg_data = pd.read_csv("ecg.csv", header=None)
ecg_data.head()
```

```python
X = ecg_data.iloc[:, :-1]
Y = ecg_data.iloc[:, -1]
X.head()
```

```python
x_train, x_test, y_train, y_test = \
        train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state= None)
```

```python
from sklearn.discriminant_analysis import StandardScaler


scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
svm_model = SVC()
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)
accuracy_score(y_test, y_pred)
kfold_scores = cross_val_score(svm_model, x_train, y_train, cv=5)
# get the average score for the cross validation
val_acc = sum(kfold_scores) / len(kfold_scores)
print("Train Accuracy: ", svm_model.score(x_train, y_train))
print("Test Accuracy: ", accuracy_score(y_test, y_pred))
print("Cross Validation Accuracy: ", val_acc)
```
